```python
import csv
import matplotlib.pyplot as plt

f = open('seoul.csv',encoding='cp949')
data = csv.reader(f)
next(data)
result = []

for row in data :
    if row[4] != '' :
        result.append(float(row[4]))

plt.hist(result, bins=100, color='r')
plt.show()

f.close()
```

```python
import csv
import matplotlib.pyplot as plt

f = open('seoul.csv',encoding='cp949')
data = csv.reader(f)
next(data)
result = []

for row in data :
    month = row[2].split('-')[1]
    if row[4] != '' :
        if month == '08' :
            result.append(float(row[4]))

plt.hist(result, bins=100, color='r')
plt.show()

f.close()
```

```python
import csv
import matplotlib.pyplot as plt

f = open('seoul.csv',encoding='cp949')
data = csv.reader(f)
next(data)

aug = []
jan = []

for row in data :
    month = row[2].split('-')[1]
    if row[4] != '' :
        if month == '08' :
            aug.append(float(row[4]))
        if month == '01' :
            jan.append(float(row[4]))

plt.hist(aug, bins=100, color='r', label='Aug')
plt.hist(jan, bins=100, color='b', label='Jan')
plt.legend()
plt.show()

f.close()
```

```python
import matplotlib.pyplot as plt
import numpy as np
import random
result = []
for i in range(13) :
    result.append(random.randint(1,1000))
    
print(sorted(result))
print("1/4: "+str(np.percentile(result,25)))
print("2/4: "+str(np.percentile(result,50)))
print("3/4: "+str(np.percentile(result,75)))

plt.boxplot(result)
plt.show()
```

```python
import matplotlib.pyplot as plt
import csv

f = open('seoul.csv', encoding = 'cp949')
data = csv.reader(f)
next(data)

month = [[],[],[],[],[],[],[],[],[],[],[],[]]

for row in data :
    if row[4] != '' :
        month[int(row[2].split('-')[1])-1].append(float(row[4]))

plt.boxplot(month)
plt.show()
```

```python
import matplotlib.pyplot as plt
import csv

f = open('seoul.csv', encoding = 'cp949')
data = csv.reader(f)
next(data)

day = []

for i in range(31) :
    day.append([])
    
for row in data :
    if row[4] != '' :
        if row[2].split('-')[1] == '08' :
            day[int(row[2].split('-')[2]) - 1].append(float(row[4]))
            
plt.style.use('ggplot')
plt.figure(figsize=(10,5), dpi=300)
plt.boxplot(day, showfliers=False)

plt.show()
```
