```python
import pandas as pd
df = pd.read_csv('age.csv', encoding='cp949',index_col=0, thousands=',')
```

```python
df = df.div(df['2019년02월_계_총인구수'], axis=0)
df.head()
del df['2019년02월_계_총인구수'], df['2019년02월_계_연령구간인구수']
```

```python
name = input('원하는 지역의 이름을 입력해주세요: ')
a = df.index.str.contains(name)

df2 = df[a]

df2
```

```python
import matplotlib.pyplot as plt

plt.rc('font', family = 'NanumGothicCoding')
df2.T.plot()
plt.show()
```

```python
import numpy as np

x = df.sub(df2.iloc[0], axis=1)
y = np.power(x,2)
z = y.sum(axis=1)
z
```

```python
i = z.sort_values().index[:5]
```

```python
df.loc[i].T.plot()
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

plt.rc('font', family = 'NanumGothicCoding')
df = pd.read_csv('age.csv', encoding='cp949', thousands=',', index_col=0)

df = df.div(df['2019년02월_계_총인구수'], axis=0)
del df['2019년02월_계_총인구수'], df['2019년02월_계_연령구간인구수']

name = input('원하는 지역의 이름을 입력해주세요: ')
a = df.index.str.contains(name)
df2 = df[a]


df.loc[np.power(df.sub(df2.iloc[0], axis=1),2).sum(axis=1).sort_values().index[:5]].T.plot()
plt.show()
```
