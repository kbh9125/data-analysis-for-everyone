```python
import matplotlib.pyplot as plt
import numpy as np

# evenly sampled time at 200ms intervals
t = np.arange(0., 5., 0.2)

# red dashes, blue squares and green triangles
plt.plot(t, t, 'r--', t, t**2, 'bs', t, t**3, 'g^')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
import random

dice_n = np.random.choice(6,10)
dice_r = []

for i in range(10) :
    dice_r.append(random.randint(1,6))

plt.subplot(121)
plt.hist(dice_n, bins = 6)
plt.subplot(122)
plt.hist(dice_r, bins = 6)

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.random.randint(-100,100,1000)
y = np.random.randint(-100,100,1000)
size = np.random.rand(100) * 100

print(size)

mask1 = abs(x) > 50
mask2 = abs(y) > 50

x = x[mask1 + mask2]
y = y[mask1 + mask2]

plt.scatter(x, y, s=size, c=x, cmap='jet', alpha=0.7)
plt.colorbar()
plt.show()
```
