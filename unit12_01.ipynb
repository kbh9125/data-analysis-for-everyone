```python
import csv

f = open('subwaytime.csv')
data = csv.reader(f)

next(data)
next(data)
result = []

for row in data :
    row[4:] = map(int, row[4:])
    result.append(row[10])
print(len(result))
print(result)

import matplotlib.pyplot as plt
plt.style.use('ggplot')
result.sort()
plt.bar(range(len(result)), result)
plt.show()
```

```python
import csv
import matplotlib.pyplot as plt

f = open('subwaytime.csv')
data = csv.reader(f)

next(data)
next(data)
mx = [0] * 24
mx_station = [''] * 24

for row in data :
    row[4:] = map(int, row[4:])
    for j in range(24) :
        a = row[j*2+4]
        if a > mx[j] :
            mx[j] = a
            mx_station[j] = row[3] + '(' + str(j+4) + ')'
for j in range(24) :
    print(j,' 시간대 -> ',mx_station[j], ' : ', mx[j])
    

plt.rc('font', family = 'NanumGothicCoding')
plt.bar(range(24),mx)
plt.xticks(range(24), mx_station, rotation=90)
plt.show()
```

```python
import csv
import matplotlib.pyplot as plt

f = open('subwaytime.csv')
data = csv.reader(f)

next(data)
next(data)
mx = [0] * 24
mx_station = [''] * 24

for row in data :
    row[4:] = map(int, row[4:])
    for j in range(24) :
        a = row[5+j*2]
        if a > mx[j] :
            mx[j] = a
            mx_station[j] = row[3] + '(' + str(j+4) + ')'
for j in range(24) :
    print(j,' 시간대 -> ',mx_station[j], ' : ', mx[j])
    

plt.rc('font', family = 'NanumGothicCoding')
plt.bar(range(24),mx)
plt.xticks(range(24), mx_station, rotation=90)
plt.show()
```

```python
import csv
import matplotlib.pyplot as plt

f = open('subwaytime.csv')
data = csv.reader(f)

next(data)
next(data)
s_in = [0] * 24
s_out = [0] * 24

for row in data :
    row[4:] = map(int, row[4:])
    for i in range(24) :
        s_in[i] += row[i*2+4]
        s_out[i] += row[i*2+5]

plt.rc('font', family = 'NanumGothicCoding')
plt.title('지하철 시단대별 승하차 인원 추이')
plt.plot(s_in, label = '승차')
plt.plot(s_out, label = '하차')
plt.legend()
plt.xticks(range(24), range(4,28))
plt.show()
```
